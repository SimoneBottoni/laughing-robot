```python
import pandas as pd
```

```python
dataset = pd.read_csv("dpc-covid19-ita-regioni_clean.csv")
```

```python
ricoverati_regioni = dataset[["data", "denominazione_regione", "ricoverati_con_sintomi"]].copy()

ricoverati_regioni.set_index("denominazione_regione", inplace=True)

ricoverati_regioni.head(100)

tot = ricoverati_regioni.pivot_table(index='denominazione_regione', columns='data', values='ricoverati_con_sintomi')
tot.head(30)
```

```python
tot.to_csv("ts_regioni_ricoverati.csv")
```
