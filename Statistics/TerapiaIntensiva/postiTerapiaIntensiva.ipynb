```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
terapie_intensive = pd.read_csv("../../Datasets/TerapiaIntensiva/Totale_posti_terapie_intensive_per_regione.csv")
```

```python
#get x label
terapie_intensive = terapie_intensive.groupby(["Descrizione Regione"], as_index=False).agg({"Totale posti letto": 'sum'})
label = terapie_intensive['Descrizione Regione'].tolist()

#Create plot
ax = terapie_intensive.plot.bar(title="Totale Posti Terapie Intensive per Regione", figsize=(16,9), rot=1)
ax.get_legend().remove()

#add x labels
plt.xticks(np.arange(len(label)), label, rotation=90)

#In order to have same vertical shifts between all bars top value
not_so_beautifull_vertical_shift = int(max(terapie_intensive["Totale posti letto"].tolist())) * 0.01

#write column values over them
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()+not_so_beautifull_vertical_shift), ha='center')    

#Show plot
plt.show()
```
