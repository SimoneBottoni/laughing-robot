```python
import pandas
import sqlalchemy

engine = sqlalchemy.create_engine("postgresql://postgres:postgres@localhost/covid19")

istat = pandas.read_csv("istat_clean.csv")
open_line_list = pandas.read_csv("COVID19_open_line_list_clean.csv")
covid19_complete = pandas.read_csv("covid_19_clean_complete_clean.csv")
france = pandas.read_csv("france_clean.csv")
gdp_health = pandas.read_csv("GDP_HEALTH_per_country_clean.csv")
andamento_nazionale = pandas.read_csv("dpc-covid19-ita-andamento-nazionale_clean.csv")
andamento_regioni = pandas.read_csv("dpc-covid19-ita-regioni_clean.csv")
andamento_province = pandas.read_csv("dpc-covid19-ita-province_clean.csv") 

istat.to_sql(
    name = "istat",
    con = engine,
    index = False,
    if_exists = "append"
)

open_line_list.to_sql(
    name = "open_line_list",
    con = engine,
    index = False,
    if_exists = "append"
)

covid19_complete.to_sql(
    name = "covid19_complete",
    con = engine,
    index = False,
    if_exists = "append"
)

france.to_sql(
    name = "france",
    con = engine,
    index = False,
    if_exists = "append"
)

gdp_health.to_sql(
    name = "gdp_health",
    con = engine,
    index = False,
    if_exists = "append"
)

andamento_nazionale.to_sql(
    name = "andamento_nazionale",
    con = engine,
    index = False,
    if_exists = "append"
)

andamento_regioni.to_sql(
    name = "andamento_regioni",
    con = engine,
    index = False,
    if_exists = "append"
)

andamento_province.to_sql(
    name = "andamento_province",
    con = engine,
    index = False,
    if_exists = "append"
)
```
